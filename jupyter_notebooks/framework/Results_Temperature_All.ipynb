## Plot Prep

```python
DayYr = np.array( [x for x in range(1, 367, 1)], dtype=np.int32 )
```

```python
DDPMin = np.minimum( DP_DryData["Tmin_C"].to_numpy(), DP_WetData["Tmin_C"].to_numpy() )
DDPMax = np.maximum( DP_DryData["Tmax_C"].to_numpy(), DP_WetData["Tmax_C"].to_numpy() )
DDPMin.shape, DDPMax.shape
```

```python
DYrList = list( DayYr )
type(DYrList)
```

```python
BackDYrList = deepcopy( DYrList )
BackDYrList = BackDYrList[::-1]
BackDYrList[:5]
```

```python
MinVals = list( DDPMin )
MaxVals = list( DDPMax )
MaxVals = MaxVals[::-1]
```

```python
DPValsList = deepcopy( MinVals )
DPValsList.extend( MaxVals )
DYrList.extend( BackDYrList )
```

```python
LP1Min = np.minimum( P1_DryData["Tmin_C"].to_numpy(), P1_WetData["Tmin_C"].to_numpy() )
LP1Max = np.maximum( P1_DryData["Tmax_C"].to_numpy(), P1_WetData["Tmax_C"].to_numpy() )
LP1Min.shape, LP1Max.shape
```

```python
MinVals = list( LP1Min )
MaxVals = list( LP1Max )
MaxVals = MaxVals[::-1]
```

```python
P1ValsList = deepcopy( MinVals )
P1ValsList.extend( MaxVals )
```

```python
LP2Min = np.minimum( P2_DryData["Tmin_C"].to_numpy(), P2_WetData["Tmin_C"].to_numpy() )
LP2Max = np.maximum( P2_DryData["Tmax_C"].to_numpy(), P2_WetData["Tmax_C"].to_numpy() )
LP2Min.shape, LP2Max.shape
```

```python
MinVals = list( LP2Min )
MaxVals = list( LP2Max )
MaxVals = MaxVals[::-1]
```

```python
P2ValsList = deepcopy( MinVals )
P2ValsList.extend( MaxVals )
```

```python
LP3Min = np.minimum( P3_DryData["Tmin_C"].to_numpy(), P3_WetData["Tmin_C"].to_numpy() )
LP3Max = np.maximum( P3_DryData["Tmax_C"].to_numpy(), P3_WetData["Tmax_C"].to_numpy() )
LP3Min.shape, LP3Max.shape
```
